```python
import requests
water_billing_url = "https://raw.githubusercontent.com/Gigomkenya/water_billing_data./refs/heads/main/MOCK_DATA%20(3).json"
```

```python
water_billing_data = requests.get(water_billing_url)
```

```python
print(water_billing_data)
```

```python
type(water_billing_data)
```

```python
water_billing_data_json= water_billing_data.json()
```

```python
water_billing_data
```

```python
type(water_billing_data.json())
```

```python
print(water_billing_data_json)
```

```python
#Convert the extracted JSON data into a well-structured DataFrame.
import pandas as pd
water_billing_data = pd.DataFrame(water_billing_data_json)
water_billing_data.head()
print(water_billing_data.head())
```

```python
#Save the final output as a well-structured CSV file.
water_billing_data.to_csv("water_billing_data.csv", index=False)
```
